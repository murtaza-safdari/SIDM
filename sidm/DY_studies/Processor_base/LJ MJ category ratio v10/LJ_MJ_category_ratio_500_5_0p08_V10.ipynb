```python
import hist.intervals
```

```python
def merge(h1, h2):
    histo = hist.Hist(*h1.axes)
    histo.values()[:] = (h1.values() + h2.values())
    
    return histo

def merge_more(h1, h2, h3):
    histo = hist.Hist(*h1.axes)
    histo.values()[:] = (h1.values() + h2.values() + h3.values())
    
    return histo
```
