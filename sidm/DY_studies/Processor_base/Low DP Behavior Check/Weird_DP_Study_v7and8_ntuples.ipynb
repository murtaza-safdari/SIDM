```python
fileset_4mu = utilities.make_fileset(samples_4mu, "llpNanoAOD_v2", max_files=5, location_cfg="signal_v8.yaml")
fileset_4mu_v7 = utilities.make_fileset(samples_4mu_v7, "llpNanoAOD_v2", max_files=5, location_cfg="signal_v7.yaml")

runner = processor.Runner(
    executor=processor.IterativeExecutor(),
    #executor=processor.FuturesExecutor(),
    schema=NanoAODSchema,
    maxchunks=5,
    skipbadfiles=True
)

channels = [
    "base",
]

p = sidm_processor.SidmProcessor(
    channels,
    ["base_plus_gen"],
)

output_4mu = runner.run(fileset_4mu, treename="Events", processor_instance=p)
out_4mu = output_4mu["out"]

output_4mu_v7 = runner.run(fileset_4mu_v7, treename="Events", processor_instance=p)
out_4mu_v7 = output_4mu_v7["out"]
```
